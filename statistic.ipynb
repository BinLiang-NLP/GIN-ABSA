```python
import pandas as pd
from IPython.core.interactiveshell import InteractiveShell
InteractiveShell.ast_node_interactivity = "all"
import os
```

```python
semeval_rest_train = './datasets/semeval14/Restaurants_Train.xml.seg'
semeval_rest_test = './datasets/semeval14/Restaurants_Test_Gold.xml.seg'
semeval_lap_train = './datasets/semeval14/Laptops_Train.xml.seg'
semeval_lap_test =  './datasets/semeval14/Laptops_Test_Gold.xml.seg'
twitter_train = './datasets/acl-14-short-data/train.raw'
twitter_test = './datasets/acl-14-short-data/test.raw'

path_list = [semeval_rest_train,semeval_rest_test,semeval_lap_train,semeval_lap_test,twitter_train,twitter_test]

def transform(path):
    with open(path,'r',encoding='utf-8',newline='\n',errors='ignore') as f:
        line = f.readlines()
    df = pd.DataFrame(columns=['text','aspect','sentiment'])
    for i in range(0,len(line),3):
        text = line[i].strip()
        aspect = line[i+1].strip()
        sentiment = line[i+2].strip()
        df = df.append(pd.Series({'text':text,'aspect':aspect,'sentiment':sentiment}),ignore_index=True)
    if not os.path.exists('dataset_pd'):
        os.mkdir('dataset_pd')
    with open('dataset_pd/'+path.split('/')[-1]+'.pd','w') as f:
        df.to_csv(f,index=False)
for i in path_list:
    transform(i)
```

```python

file_path = './datasets/semeval14/Restaurants_Train.xml.seg'
def statistic_aspect(path):
    with open(path,'r',encoding='utf-8',newline='\n',errors='ignore') as f:
        line = f.readlines()
    df = pd.DataFrame(columns=['text','aspect','sentiment'])
    for i in range(0,len(line),3):
        text = line[i].strip()
        aspect = line[i+1].strip()
        sentiment = line[i+2].strip()
        df = df.append(pd.Series({'text':text,'aspect':aspect,'sentiment':sentiment}),ignore_index=True)
    df.head()
    df.shape
    r = df['aspect'].map(lambda x:x.lower()).value_counts()
    print(r)
    return r
```
